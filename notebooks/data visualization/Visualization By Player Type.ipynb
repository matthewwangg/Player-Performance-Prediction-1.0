```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("players.csv")
# Set the style for seaborn plots
sns.set(style="whitegrid")
```

```python
# Count of players by position
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='position')
plt.title("Count of Players by Position")
plt.xlabel("Position")
plt.ylabel("Count")
plt.show()
```

```python
# Goals by position (bar plot to show average goals by position)
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='position', y='goals_scored')
plt.title("Average Goals by Position")
plt.xlabel("Position")
plt.ylabel("Average Goals")
plt.show()
```

```python
# Assists by position (scatter plot to show distribution of assists)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='position', y='assists', size='assists', legend=False)
plt.title("Assists Distribution by Position")
plt.xlabel("Position")
plt.ylabel("Assists")
plt.show()
```

```python
# Bonus points by position (line plot to see trends in playing time)
plt.figure(figsize=(8, 6))
sns.lineplot(data=df, x='position', y='bonus', marker='o')
plt.title("Bonus Points by Position")
plt.xlabel("Position")
plt.ylabel("Minutes Played")
plt.show()
```

```python
plt.figure(figsize=(6, 4))  # Adjust the figure size
sns.barplot(data=df, x='position', y='starts_per_90')  # Create a bar plot
plt.title("Starts per 90 by Position")  # Title of the plot
plt.xlabel("Position")  # Label for the x-axis
plt.ylabel("Starts per 90")  # Label for the y-axis
plt.show()  # Display the plot
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='position', y='expected_goal_involvements')
plt.title("Expected Goal Involvements by Position")
plt.xlabel("Position")
plt.ylabel("Expected Goal Involvements")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='position', y='points_per_game')
plt.title("Points per Game by Position")
plt.xlabel("Position")
plt.ylabel("Points per Game")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='position', y='creativity_rank_type')
plt.title("Creativity Rank Type by Position")
plt.xlabel("Position")
plt.ylabel("Creativity Rank Type")
plt.show()
```
